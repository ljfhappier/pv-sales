```python
train_sales_data=pd.read_csv('data/train_sales_data.csv')
train_sales_data.head(2)
```

```python
train_sales_data.info()
```

```python
train_data_2016=train_sales_data[train_sales_data['regYear']==2016]
train_data_2017=train_sales_data[train_sales_data['regYear']==2017]
train_data_tb=pd.merge(train_data_2017,train_data_2016,on=['province','adcode','model','bodyType','regMonth'],\
                        how='left',suffixes=('','_tb'))
train_data_tb.drop('regYear_tb',inplace=True,axis=1)
train_data_tb.head()
```

```python
train_data_2017_hb=train_sales_data[train_sales_data['regYear']==2017]
train_data_2017_hb['regMonth']=train_data_2017_hb['regMonth']+1
#train_data_2017_hb.head()
train_data_hb=pd.merge(train_data_2017,train_data_2017_hb,on=['province','adcode','model','bodyType','regYear','regMonth'],\
                        how='left',suffixes=('','_hb'))
train_data_hb.head()
```

```python
train_data_201612=train_data_2016[train_data_2016['regMonth']==12]
cols_201612=['province','adcode','model','bodyType','salesVolume']
train_data_201612=train_data_201612[cols_201612]
train_data_hb=pd.merge(train_data_hb,train_data_201612,on=['province','adcode','model','bodyType'],\
                         how='left',suffixes=('','_bc'))
lsindex_hb=train_data_hb['salesVolume_hb'].isnull()
train_data_hb.loc[lsindex_hb,'salesVolume_hb']=train_data_hb.loc[lsindex_hb,'salesVolume_bc']
train_data_hb.drop('salesVolume_bc',inplace=True,axis=1)
train_data_hb.head()
```

```python
cols=['province','adcode','model','bodyType','regMonth','salesVolume']
train_data_2017_1=train_data_2017[cols]
train_data_2017_1['regMonth']=train_data_2017_1['regMonth']-1
train_data_y1=pd.merge(train_data_2017,train_data_2017_1,on=['province','adcode','model','regMonth','bodyType'],\
                       how='left',suffixes=('','_y1'))
train_data_y1.head()
```

```python
cols=['province','adcode','model','bodyType','regMonth','salesVolume']
train_data_2017_2=train_data_2017[cols]
train_data_2017_2['regMonth']=train_data_2017_2['regMonth']-2
train_data_y2=pd.merge(train_data_2017,train_data_2017_2,on=['province','adcode','model','regMonth','bodyType'],\
                       how='left',suffixes=('','_y2'))
train_data_y2.head()
```

```python
cols=['province','adcode','model','bodyType','regMonth','salesVolume']
train_data_2017_3=train_data_2017[cols]
train_data_2017_3['regMonth']=train_data_2017_3['regMonth']-3
train_data_y3=pd.merge(train_data_2017,train_data_2017_3,on=['province','adcode','model','regMonth','bodyType'],\
                       how='left',suffixes=('','_y3'))
train_data_y3.head()
```

```python
train_data_y3.loc[train_data_y3['salesVolume_y3'].isnull(),'regMonth'].value_counts()
```

```python
on_cols=['province','adcode','model','bodyType','regYear','regMonth','salesVolume']
train_data1=pd.merge(train_data_tb,train_data_hb,on=on_cols)
train_data2=pd.merge(train_data1,train_data_y1,on=on_cols);train_data2.drop('salesVolume',axis=1,inplace=True)
train_data3=pd.merge(train_data1,train_data_y2,on=on_cols);train_data3.drop('salesVolume',axis=1,inplace=True)
train_data4=pd.merge(train_data1,train_data_y3,on=on_cols);train_data4.drop('salesVolume',axis=1,inplace=True)
train_data4.head(1).T
```
